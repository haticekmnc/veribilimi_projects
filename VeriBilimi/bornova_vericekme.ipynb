```python
import requests

bornova = requests.get("https://izmirbal.meb.k12.tr/tema/index.php")
```

```python
#sayfa başarılı bir şekilde indirilmiş mii
#200 döndü başarılı

bornova.status_code
```

```python
#sayfa içeriği

bornova.content
```

```python
#beautifulSoup

from bs4 import BeautifulSoup

soup = BeautifulSoup(bornova.content, 'html.parser')
soup
```

```python
#prettify işlevi

print(soup.prettify())
```

```python
#başlık bilgisini alalım

baslik = soup.find(id="okul")

print(baslik.get_text())
```

```python
#bağlantılar

bag = soup.find(id="dik_menu")

print(bag.get_text())
```

```python
#bağlantılar alt başlıkları

baglantilar = bag.find_all(class_="list-group-item")

for b in baglantilar:
  print(b.text)
```

```python
b_list = [b.text for b in baglantilar]
b_list
```

**Haberler sayfasında çalışıyoruz**


```python
import requests

haberler = requests.get("https://izmirbal.meb.k12.tr/tema/icerik.php?KATEGORINO=96121")
```

```python
haberler.status_code
```

```python
from bs4 import BeautifulSoup

soup_haber = BeautifulSoup(haberler.content, 'html.parser')
```

```python
soup_haber.prettify()
```

```python
#haberler sayfası

icerik = soup_haber.find(id="liste")
 #haber başlıkları

haber_basliklari = icerik.find_all(class_="liste_baslik")
```

```python
#ilk haber başlığını yazdıralım

print(haber_basliklari[0].get_text())
```

```python
#tarih bilgileri

tarih_bil = icerik.find_all(class_="tarih")
```

```python
#ilk tarih bilgisini yazdıralım

print(tarih_bil[0].get_text())
```

```python
#gün, ay ve yıl

gun = icerik.find_all(class_="gun")
ay = icerik.find_all(class_="ay")
yil = icerik.find_all(class_="yil")
```

```python
#ilk haber gün bilgisi

print(gun[0].get_text())
```

```python
#haber başlıklarını listeye atma

print(len(haber_basliklari))
```

```python
hb = []

for i in range(len(haber_basliklari)):
  hb.append(haber_basliklari[i].get_text())
```

```python
hb
```

```python
#gün listesi

gunler= []

for i in range(len(gun)):
  gunler.append(gun[i].get_text())
```

```python
gunler
```

```python
#ay bilgileri

aylar = []

for i in range(len(ay)):
  aylar.append(ay[i].get_text())
```

```python
aylar
```

```python
#yıl bilgileri

yillar = []

for i in range(len(yil)):
  yillar.append(yil[i].get_text())
```

```python
yillar
```

***Pandas dataframe çalışması***

```python
import pandas as pd

haber_df = pd.DataFrame({"haber":hb, "gün":gunler, "ay":aylar, "yıl":yillar})
```

```python
haber_df
```

```python
# "\n" temizleme

haber_df["haber"] = haber_df["haber"].str.replace("\n", "")

haber_df
```

```python
# bütün harfleri küçüğe döndürme

haber_df["haber"] = haber_df["haber"].str.lower()

haber_df

```
